```python
join_orders_cust = pd.merge(left = orders, right = Customer,on = "CustomerID", how = "inner")
join_orders_cust
```

```python
revenue_city = join_orders_cust.groupby("City").agg({"OrderID":"count",'TotalAmount':'sum'})
revenue_city
```

```python
status = join_orders_cust.groupby("DeliveryStatus").agg({"TotalAmount":"sum"})
status
```

```python
sns.barplot(x = status.index, y = status.TotalAmount, data = status, color = "orange")
plt.xlabel("Delivery Status")
plt.ylabel("Total Amount")
plt.title("Total Amount by Delivery Status")
plt.show()
```

```python
plt.pie(status["TotalAmount"], labels= status.index, colors=["red","green"], autopct = "%1.2f%%")
plt.title("Total Amount by Delivery Status")
plt.show()
```
